```python
# protected variable
class ParentClass:
    def __init__(self):
        self.public_var = "I'm a public variable"
        self._protected_var = "I'm a protected variable in the parent class"

    def get_protected_var(self):
        return self._protected_var

    def set_protected_var(self, value):
        self._protected_var = value

class ChildClass(ParentClass):
    def __init__(self):
        super().__init__()
        self.child_var = "I'm a variable in the child class"

    def get_protected_var_from_parent(self):
        return self._protected_var
```

```python
obj = ChildClass()
print(obj.public_var)                    # Output: I'm a public variable
print(obj.get_protected_var())           # Output: I'm a protected variable in the parent class
print(obj.get_protected_var_from_parent())# Output: I'm a protected variable in the parent class

obj._protected_var = "New value"        # This directly modifies the protected variable in the parent class
print(obj.get_protected_var())           # Output: New value

obj.set_protected_var("New value")      # This also sets a new value for the protected variable in the parent class
print(obj.get_protected_var())           # Output: New value
```
