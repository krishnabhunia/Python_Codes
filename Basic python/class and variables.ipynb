```python
# Access of private variable via class name

class Test:
    __cvar = "this is class variable"
    def __init__(self):
        self.__var = "this is instance variable inside the initializer"
        print(self.__var)
        
    @classmethod
    def getCVar(cls):
        return Test.__cvar

    def getVar(self):
        return self.__var

t = Test() # creating instance
print("calling class variable : ",Test.getCVar())
print("calling instance variable : ",t.getVar())
print("direct calling private instance variable : ",t._Test__var)
print("direct calling private class variable : ",t._Test__cvar)

t._Test__var = 100 # updating instance private variable
Test._Test__cvar = 999 # updating class private varible
t.__cvar = "ok"
print("After assigning")
print("calling class variable : ",Test.getCVar())
print("calling instance variable : ",t.getVar())
print("direct calling private instance variable : ",t._Test__var)
print("direct calling private class variable : ",Test._Test__cvar)

Test._Test__var = 555
t._Test__cvar = 666
print("After assigning")
print("calling class variable : ",Test.getCVar())
print("calling instance variable : ",t.getVar())
print("direct calling private instance variable : ",t._Test__var)
print("direct calling private class variable : ",Test._Test__cvar)

print("Fetching")
print(Test._Test__var)
print(t._Test__cvar)
```

```python
# Operator Overloading

class OO():
    def __init__(self,data):
        self.d1 = data

    def __add__(self,other):
        self.res = self.d1 + other.d1
        return f"This addition of objects is {self.res}"
    
    def __and__(self, other):
        self.res = self.d1 & other.d1
        return f"Result of and operation :{self.res}"
    
    def __ge__(self, other):
        if self.d1 > other.d1:
            return "True greater or equal executed"
        else:
            return "False greater or equal executed"
        
    def __gt__(self, other):
        if self.d1 >= other.d1:
            return "True greater executed"
        else:
            return "False greater executed"
    
    def __le__(self, other):    
        pass
    
    def __sub__(self, other):
        pass
    
    def __float__(self):
        return "__float__ has been called"
    

f = OO(10) #  0101
s = OO(20) # 10100
# res and    00100
print(f+s)
print(f&s)
print(f>=s)
print(f>s)
print(f<s)
print(f<=s)
```

```python
# Duck Typing
class Horse():
    def walk(self):
        print('thapk thapk')

class Duck():
    def walk(self):
        print('quack quack')

class Cat():
    def talk(self):
        print('meow meow')

def myfun(obj):
    if hasattr(obj,'walk'):
        obj.walk()
    if hasattr(obj,'talk'):
        obj.talk()

h=Horse()
d=Duck()
c=Cat()
myfun(h)
myfun(d)
myfun(c)
```

```python
# Inheritance with protected variable

class A:
    _pcva = "this is protected class variable of A"
    def __init__(self):
        super().__init__()
        self._pva = "this is protected instance variable of class A"
        print("This is initializer of class A")
        
    def show(self):
        print("ShowClass Start")
        print("Value of protected instance variable of class A:",self._pva)
        print("Value of protected class variable of class A:",self._pcva)
        
        # print("Value of protected instance variable of class B:",self._pvb) Inheritance Problem
        # print("Value of protected class variable of class B:",self._pcvb) # Inheritance Problem
        print("ShowClass Stop")

        
    @classmethod
    def showclass(cls):
        print("ShowClass Start")
        # print("Value of protected instance variable of class A:",cls._pva) # AttributeError: type object 'A' has no attribute '_pva'
        print("Value of protected class variable of class A:",cls._pcva)
        print("ShowClass Stop")

class B(A):
    _pcvb = "this is protected class variable of B"
    def __init__(self):
        super().__init__()
        self._pvb = "this is protected instance variable of class B"
        print("This is initializer of class B")
        
    def show(self):
        print("Show Start")
        print("Value of protected instance variable of class B:",self._pvb)
        print("Value of protected class variable of class B:",self._pcvb)
        
        print("Value of protected instance variable of class A:",self._pva)
        print("Value of protected class variable of class A:",self._pcva)
        print("Show Stop")
        
        
    @classmethod
    def showclass(cls):
        print("ShowClass Start")
        # print("Value of protected instance variable of class A:",cls._pva) # AttributeError: type object 'A' has no attribute '_pva'
        print("Value of protected class variable of class B:",cls._pcvb)
        
        print("Value of protected class variable of class A:",cls._pcva)
        print("ShowClass Stop")
        
a = A()
# b = B()
a.show()
A.showclass()

# b.show()
# b.showclass()

```

```python
from multipledispatch import dispatch
# the most efficient one to use method overloading

# passing one parameter
@dispatch(str,str)
def p1(first, second):
	result = int(first)*int(second)
	print("str,str",result)
    
@dispatch(int,int)
def p1(first, second):
	result = first*second
	print("int,int",result)

# passing two parameters
@dispatch(str,str,str)
def p1(first, second, third):
	result = first * second * third
	print("str,str,str",result)

# you can also pass data type of any value as per requirement
@dispatch(float, float, float)
def p1(first, second, third):
	result = first * second * third
	print("float, float, float",result)
    
@dispatch(int,int)
def p1(first, second):
	result = first * second
	print("int,int",result)
    
@dispatch(int,int,int)
def p1(first, second, third):
	result = first * second * third
	print("int,int,int",result)
    
@dispatch(float,float)
def p1(first, second):
	result = first * second
	print("float,float",result)
    
# print(help(product))

# calling product method with 2 arguments
p1("2", "3") # this will give output of 6

# calling product method with 3 arguments but all int
p1(2, 3, 2) # this will give output of 12

# calling product method with 3 arguments but all float
p1(2.2, 3.4, 2.3) # this will give output of 17.985999999999997

# calling product method with 3 arguments but all float
p1(2.2, 3.4) # this will give output of 7.48
```
